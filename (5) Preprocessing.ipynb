## Tahap 1

```python
DKI1 = DKI1.drop(columns=["Tanggal", "Wilayah", "Max", "Critical"])
DKI2 = DKI2.drop(columns=["Tanggal", "Wilayah", "Max", "Critical"])
DKI3 = DKI3.drop(columns=["Tanggal", "Wilayah", "Max", "Critical"])
DKI4 = DKI4.drop(columns=["Tanggal", "Wilayah", "Max", "Critical"])
DKI5 = DKI5.drop(columns=["Tanggal", "Wilayah", "Max", "Critical"])
```

## Tahap 5
melakukan pengecekan label "sangat tidak sehat" karenakan datanya terlalu sedikit sehingga akan berpengaruh kepada klasifikasi(data tidak seimbang)

```python
#DKI5.drop(DKI5[DKI5['Kategori'] == 'sangat tidak sehat'].index,inplace= True)
DKI5[DKI5['Kategori'] == 'sangat tidak sehat'].count()
```

```python
def filter_category(df, category, start=0, end=100):
    return df.loc[df['Kategori'] == category].iloc[start:end]
    
def category_not_health(df, start=0, end=100):
    return filter_category(df, 'tidak sehat', start, end)
    
def category_intermediate(df, start=0, end=100):
    return filter_category(df, 'sedang', start, end)

def category_health(df, start=0, end=100):
    return filter_category(df, 'baik', start, end)

def category_very_not_health(df, start=0, end=100):
    return filter_category(df, 'sangat tidak sehat', start, end)

def sample_not_health():
    return pd.concat([
        category_not_health(DKI1),
        category_not_health(DKI2, end=135),
        category_not_health(DKI3),
        category_not_health(DKI4),
        category_not_health(DKI5),
    ], ignore_index=True)

def sample_intermediate():
    return pd.concat([
        category_intermediate(DKI1),
        category_intermediate(DKI2),
        category_intermediate(DKI3),
        category_intermediate(DKI4),
        category_intermediate(DKI5),
    ], ignore_index=True)

def sample_health():
    return pd.concat([
        category_health(DKI1),
        category_health(DKI2),
        category_health(DKI3),
        category_health(DKI4),
        category_health(DKI5),
    ], ignore_index=True)

def sample_very_not_health():
    return pd.concat([
        category_very_not_health(DKI5)
    ], ignore_index=True)
```

```python
not_health = sample_not_health()
not_health
```

```python
intermediate = sample_intermediate()
intermediate
```

```python
health = sample_health()
health
```

```python
very_not_health = sample_very_not_health()
very_not_health
```

```python
imbalance_dataset = pd.concat([very_not_health, not_health, intermediate, health], ignore_index=True)
category_very_not_health(imbalance_dataset)
```

```python
imbalance_X=imbalance_dataset.drop(columns=["Kategori"])
imbalance_y = imbalance_dataset["Kategori"]
imbalance_y.loc[imbalance_y == 'sangat tidak sehat'].count()
```

## Tahap 6

```python

oversampled = SMOTE()
balanced_X, balanced_y = oversampled.fit_resample(imbalance_X, imbalance_y)
balanced_y.loc[balanced_y == 'sangat tidak sehat'].count()
```
